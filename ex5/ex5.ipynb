```python
from firebase import firebase
import json as json
from google.colab import drive
#mount the google drive
drive.mount('/content/drive')
#open the connection to database
firebaseConn = firebase.FirebaseApplication('https://tirgol5-c0fed-default-rtdb.europe-west1.firebasedatabase.app', None)

#read Student.json from google
#file path (change it accordngly to where you chose to save the json file in the google drive)
filePath= "/content/drive/My Drive/Colab_Notebooks1/Students.json"
with open(filePath,'r') as studentsFile:
  studentsData=json.load(studentsFile)

#save StudentsData in the firebase database
result = firebaseConn.post('/tirgol5/',studentsData)


```

```python
import ipywidgets as widgets
from IPython.display import display
studentDataFromDB = firebaseConn.get('/tirgol5', None)

# Use a set to store unique student names
student_names = set()
print(studentDataFromDB.items())
# Extract student names from the dictionary, handling nested data
for key, value in studentDataFromDB.items():
    if isinstance(value, list):
        for student in value:
            if isinstance(student, dict) and 'name' in student:
                student_names.add(student['name'])
    elif isinstance(value, dict) and 'name' in value:
        student_names.add(value['name'])

# Convert set back to a sorted list (if needed)
student_names = sorted(list(student_names))

# Create a dropdown menu for student names
name_dropdown = widgets.Dropdown(
    options=student_names,
    description='Select Student:',
    style={'description_width': 'initial'}
)

# Create text areas for displaying information
info_name = widgets.Text(
    value='',
    description='Name:',
    disabled=True,
    style={'description_width': 'initial'}
)

info_email = widgets.Text(
    value='',
    description='Email:',
    disabled=True,
    style={'description_width': 'initial'}
)

info_courses = widgets.Textarea(
    value='',
    description='Courses:',
    disabled=True,
    style={'description_width': 'initial'}
)

#The "info_link" widget's description was updated to 'something interesting' from 'Favorite Show'
#due to the fact our json file have an empty filed for favorate show
info_link = widgets.HTML(
    value='',
    description='Something interesting:',
    disabled=True,
    style={'description_width': 'initial'}
)


# Function to update information when a student is selected
def update_info(change):
    selected_name = change['new']
    for key, value in studentDataFromDB.items():
      if isinstance(value, list):
        for student in value:
          if isinstance(student, dict) and 'name' in student:
            if selected_name == student['name']:
              student_info = student
    if student_info:
        # Update the widgets with the student's information
        info_name.value = student_info['name']
        info_email.value = student_info['mail']
        info_courses.value = student_info['Courses']
        info_link.value = f'<a href="{student_info.get("something interesting", "")}" target="_blank">{student_info.get("something interesting", "")}</a>'


# Assign the function to the dropdown's observe event
name_dropdown.observe(update_info, names='value')

# Display the widgets
widgets.VBox([name_dropdown, info_name, info_email, info_courses, info_link])
```
